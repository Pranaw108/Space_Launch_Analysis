# Business Scenario

A commercial satellite launch company plans to deploy more than 500 satellites over the next five years for Earth observation, communications, and climate-monitoring customers.

The company works with multiple launch providers and launch sites worldwide. Leadership needs a data-driven strategy to:

* Reduce launch costs
* Improve mission success rates
* Optimize payload utilization
* Select the best launch sites
* Improve annual deployment schedules

The executive team wants a dashboard that helps compare launch providers, rocket families, and launch locations to support future launch procurement decisions.

As a Data Analyst, your role is to transform historical launch data into actionable business insights and recommendations.
